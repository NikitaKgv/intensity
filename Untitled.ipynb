```python
hdu1name='11.fits'
hdu2name='22.fits'
hdu3name='33.fits'
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
from astropy.io import fits
from numpy import *
```

```python
hdulist1 = fits.open(hdu1name)
hdulist2 = fits.open(hdu2name)
hdulist3 = fits.open(hdu3name)
hdu1 = hdulist1[0]
hdu2 = hdulist2[0]
hdu3 = hdulist3[0]
```

```python
fwhm_to_sigma = 1./(8*np.log(2))**0.5
Num_pix1 = 2*math.pi*hdu1.header["BMAJ"]*hdu1.header["BMIN"]*3600000**2*fwhm_to_sigma**2/(0.06**2)
print(np.sum(hdu1.data)/Num_pix1)
Num_pix2 = 2*math.pi*hdu2.header["BMAJ"]*hdu2.header["BMIN"]*3600000**2*fwhm_to_sigma**2/(0.12**2)
print(np.sum(hdu2.data)/Num_pix2)
Num_pix3 = 2*math.pi*hdu3.header["BMAJ"]*hdu3.header["BMIN"]*3600000**2*fwhm_to_sigma**2/(0.24**2)
print(np.sum(hdu3.data)/Num_pix3)
```

```python
fwhm_to_sigma = 1./(8*np.log(2))**0.5
Num_pix1 = 2*math.pi*hdu1.header["BMAJ"]*hdu1.header["BMIN"]*3600000**2*fwhm_to_sigma**2/(0.06**2)
print(np.sum(hdu1.data[0,0,126:132,124:130])/Num_pix1)
Num_pix2 = 2*math.pi*hdu2.header["BMAJ"]*hdu2.header["BMIN"]*3600000**2*fwhm_to_sigma**2/(0.12**2)
print(np.sum(hdu2.data[0,0,63:66,62:65])/Num_pix2)
Num_pix3 = 2*math.pi*hdu3.header["BMAJ"]*hdu3.header["BMIN"]*3600000**2*fwhm_to_sigma**2/(0.24**2)
print(np.sum(hdu3.data[0,0,32,31])/Num_pix3)
```
